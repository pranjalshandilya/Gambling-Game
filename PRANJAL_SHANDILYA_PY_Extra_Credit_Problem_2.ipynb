```python
"""
Created on Mon Aug 31 19:26:12 2020

@author: Pranjal Shandilya
"""

import random                                                    


counter = 0                                                       # count of number of turns set to be 0.
balance = 200                                                     # initial balance with the customer
e_wallet = 5000                                                   # initial wallet balance
added_money = 0                                                    

while balance <= 500:                                             # checking balance with customer in start
    print("Your current balance is: ", balance)                                          
    print('You still do not have enough money to start playing.')
    added_money = int(input('Enter the money to be added: '))     # adding money to the customer balance from e-wallet if found less than 500
    balance = balance + added_money                                
    e_wallet = e_wallet-added_money                             
    print("Current Balance:INR{}".format(balance))
    print("Current e-wallet Balance:INR{}".format(e_wallet))
    if balance >= 500:                                             # again checking for the balance with the customer
        print("Now, you are ready to play the game.")
        break

game = ' '                                                         # initiating game string for yes/no
n = random.randint(0,9)                                            # generating random number from system between 0 to 9
num = eval(input("Guess the number: "))                            # taking input from user as 'Yes'
while num != n:                                                    
    if (num >= 10 or num < 0):                                     # checking number between is between 0 to 9
        print("Please enter a number between 0 & 9")

    if num > n:                                                    # Telling the user about the guess 
        print("Too high")
    else:
        print("Too low")
    counter+=1                                                     
    if counter == 2 & num==n:                                      # In second chance if user guesses right he gets Rs 1000
        print("You have just won INR 1,000")
        balance = balance + 1000
        print("Your balance: ", balance)
    if counter == 3:                                               # In third chance if user guesses wrong Rs. 500 gets deducted
        print("Sorry you lost!")
        balance = balance - 500
        print("Your balance: ", balance)
        break

    n = random.randint(0,9)
    num = eval(input("Guess the number: "))
    if n == num:                                                    # If user guesses right in 2nd and 3rd chance he gets nothing but 
        print("You got it right! Congo!")                           
        


while n == num:                                                     # If he guesses in first attempt then he gets Rs. 5000
        print("You got it right! Congo!")
        print("You have just won INR 5,000")
        balance = balance + 5000
        print("Your balance: ", balance)
        break


        
counter = 0                                                          # resetting counter to 0 for 2nd playtime
game = str(input("Want to play another game (Yes/No): "))            # taking input as Yes/No for 2nd playtime
n = random.randint(0,9)
while game == 'Yes':
    if counter == 3:                                                 # When the counter reaches to 3 loops, the loop breaks
        break
    if balance >= 500:                                               # Checks for minimum balance of Rs. 500
        num = eval(input("Guess the number: "))
        while num != n:                                              # Loop starts again to compare user input with random generation
            if (num >= 10 or num < 0):
                print("Please enter a number between 0 & 9")
        
            if num > n:
                print("Too high")
            else:
                print("Too low")
            counter+=1
            if counter == 2 & num==n:
                print("You have just won INR 1,000")
                balance = balance + 1000
                print("Your balance: ", balance)
            if counter == 3:
                print("Sorry you lost!")
                balance = balance - 500
                print("Your balance: ", balance)
                break
        
            n = random.randint(0,9)
            num = eval(input("Guess the number: "))
            
            if n == num:
                print("You got it right! Congo!")
                if n==num & counter == 1: 
                    print("You have just won INR 5,000")
                    balance = balance + 5000
                    print("Your balance: ", balance)
                    break
                
            
    else:                                                               # If balance is lesser than Rs. 500, this control will ask user for money addition
        print('You still do not have enough money to start playing.')           
        added_money = int(input('Enter the money to be added: '))
        balance = balance + added_money
        e_wallet = e_wallet-added_money 
        print("Current Balance:INR{}".format(balance))
        print("Current e-wallet Balance:INR{}".format(e_wallet))
        continue


while game == 'No':                                                      # If user enters 'No', Then program exits
    print('Have a wonderful day!')
    break
```
